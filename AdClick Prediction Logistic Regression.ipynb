# Logistic Regression



** Split the data into training set and testing set using train_test_split**

```python
from sklearn.model_selection import train_test_split
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=2323)
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

** Train and fit a logistic regression model on the training set.**

```python
lgst = LogisticRegression()
```

```python
lgst.fit(x_train,y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
pred = lgst.predict(x_test)
pred
```

** Create a classification report for the model.**
** A scenario in which you have huge benefit from a true positive, not too high cost of a false positive then recall is a good measure. If we have high cost of false positive, hopefully a low cost of false negative precision is the measure.

**In our case true positive results in website traffic and revenue whereas false positive has a low cost.

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,pred)
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax,  fmt='g'); #annot=True to annotate cells
 atrix'); 
```

```python
#Accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
#prediction score
lgst.score(x_test, y_test)
```

```python
#ROC Curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
fpr, tpr, thresholds = roc_curve(y_test,pred)
auc = roc_auc_score(y_test,pred)
auc
```

#### If class labels are mostly negative or mostly positive, a classifier that always outputs 0 or 1, respectively, will achieve high accuracy. In terms of AUC it will score 0.5
